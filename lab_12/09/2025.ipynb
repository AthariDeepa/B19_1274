```python
from typing_extensions import dataclass_transform
import pandas as pd
import numpy as np
#creating a dataset with missing values
data={'Name':['Ravi','Neha','Kiran','Sita'],'Age':[25,np.nan,30,28],'Marks':[85,90,np.nan,92]}
df=pd.DataFrame(data)
print("original DataFrame:")
print(df)
#detecting missing values
print("\ncheck for Missing values:")
print(df.isnull())
```

```python
#drop rows with missing values
df_dropped=df.dropna()
print("After Dropping  Rows with NaN:")
print(df_dropped)
```

```python
#fill missing values with a constant
df_filled=df.fillna(0)
print("After Filling Missing Values:")
print(df_filled)
```

```python
data={'ID':['1','2','3','4'],'Marks':['85','90','78','92']}
df=pd.DataFrame(data)
print(df)
```

```python
#applying transformation-add 5 bonus marks
df['Marks'] = df['Marks'].astype(float) # Convert 'Marks' to float
df['Marks']=df['Marks'].apply(lambda X:X+5)
print("After Adding Bonus Marks:")
print(df)
```

```python
#Normalization of Marks
df['Marks']=df['Marks']/df['Marks'].max()-df['Marks'].min()
print("After Normalization:")
print(df)
```
