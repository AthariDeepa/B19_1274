```python
import pandas as pd

# Create student data
student_data = {
    'Roll No': [101, 102, 103, 104],
    'Name': ['Ravi', 'Neha', 'Kiran', 'Sita'],
    'm1': [85, 90, 78, 92],
    'm2': [88, 95, 80, 85],
    'm3': [75, 89, 82, 90]
}

# Create DataFrame
df = pd.DataFrame(student_data)

# Calculate Total, Average, Result
df['Total'] = df[['m1', 'm2', 'm3']].sum(axis=1)
df['Average'] = df['Total'] / 3
df['Result'] = df[['m1', 'm2', 'm3']].apply(lambda x: 'Fail' if any(m < 35 for m in x) else 'Pass', axis=1)

# Print DataFrame
print("Student Data:")
print(df)

# Save to Excel
df.to_excel("student_data.xlsx", index=False)
print("\nExcel file 'student_data.xlsx' has been created successfully.")
```
